```python
#Logan Fehn
#Finding how much money was spent in each department
departmentList=[]
departmentCost=[]
i=0
for line in file:
    #splitting up each line into different variables
    t,ip,email,st,pN,tNotif,card,cost,dept,comp=line.split(",",9)
    #only want to append the dept once not multiple times
    if dept not in departmentList and i!=0:
        #getting the lists set up - parallel lists
        departmentList.append(dept)
        departmentCost.append(0)
    #adding the cost to the according list index based on the deptList
    if i!=0:
        departmentCost[departmentList.index(dept)]+=float(cost)
    i+=1
for i in range(len(departmentList)):
    print(f'Dept: {departmentList[i]} - Cost: ${round(departmentCost[i],2)}')
#end
```

```python
#Logan Fehn
#Finding how much was spent in each state
stateList=[]
stateCost=[]
i=0
for line in file:
    #splitting each line
    t,ip,email,st,pN,tNotif,card,cost,dept,comp=line.split(",",9)
    #almost exact same as the dept one
    #setting up parallel lists
    if st not in stateList and i!=0:
        stateList.append(st)
        stateCost.append(0)
    #then adding it based off of index
    if i!=0:
        stateCost[stateList.index(st)]+=float(cost)
    i+=1
for i in range(len(stateList)):
    print(f'State: {stateList[i]} - Cost: ${round(stateCost[i],2)}')
#end
```

```python
#Tyler Joest
#Finding how many purchases were made in each state

representedStates=[]
statePurchases=[]

for line in file:
    time,ip,email,state,phoneNumber,textNotification,card,cost,dept,junk=line.split(",",9)
    if state not in representedStates and state!="state":
        representedStates.append(state)
        statePurchases.append(0)
    if state!="state":
        statePurchases[representedStates.index(state)]+=1

for i in range(len(representedStates)):
    print(f'State: {representedStates[i]} had {statePurchases[representedStates.index(representedStates[i])]} purchases')
#end
```

```python
#Tyler Joest
#Finding how many student purchased items

eduEmails=[]
domains=[]

for line in file:
    time,ip,email,state,phoneNumber,textNotification,card,cost,dept,junk=line.split(",",9)
    email = email.split('.')
    if email[len(email)-1] not in domains:
        domains.append(email[len(email) - 1])
    if email[len(email) - 1] == "edu":
        eduEmails.append(email[0])
        
print(f'{len(eduEmails)} students purchased items')
#end
```
